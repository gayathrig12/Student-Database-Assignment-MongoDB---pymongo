 Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.

```python
x = records.aggregate(
[{
    "$project" :
    {
        "name":1,
        "_id" : 0,
        "result":{
            "$cond":
                    {"if": {"$and" : [{"$gte" : ["$Exam", 40]}, {"$gte" : ["$Quiz", 40]}, {"$gte" : ["$HomeWork", 40]}]},
                    "then" :"pass",
                    "else":"fail"
                 }
                }
    }
}])
for i in x:
    print(i)
```
